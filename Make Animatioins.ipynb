```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
color1 = np.array([255, 0, 0])  # red
color2 = np.array([255, 255, 0])  # yellow
```

```python
for i in range(1, 43):
  num = str(i).zfill(4)
  img = np.array(Image.open(f'Worm 1/2022.02.11_ERT106_OP50_D2_40X_40_1_2064-1_Z{num}.tif'))
  
  trimap = np.zeros(img.shape, dtype=np.uint8)

  # Map color1 to 255, color2 to 128, and everything else to 0
  trimap[(img == color1).all(axis=-1)] = 255
  trimap[(img == color2).all(axis=-1)] = 128
  Image.fromarray(trimap).save(f'Worm 1_trimaps/2022.02.11_ERT106_OP50_D2_40X_40_1_2064-1_Z{num}_trimap.tif')
```

```python
trimaps = []

for i in range(1, 43):
  # read images as grayscale
  img = np.array(Image.open(f'Worm 1/2022.02.11_ERT106_OP50_D2_40X_40_1_2064-1_Z{str(i).zfill(4)}_trimap.tif').convert('L'))
  trimaps.append(img)
```

```python
actual_images = []
for i in range(1, 43):
  num = str(i).zfill(4)
  img = np.array(Image.open(f'Actual files/2022.02.11_ERT106_OP50_D2_40X_40_1_2064-1_Z{num}.tif'))
  actual_images.append(img)
```

```python
plt.imshow(actual_images[0], cmap='gray')
```

```python
# Read each image file and add it to a list
images = actual_images

# Save the images as an animation
# imageio.mimsave('trimaps.gif', images, duration=1,loop=True)  # fps specifies the frames per second
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Assuming 'images' is your list of images
fig = plt.figure()

# A function to update the figure
def update(i):
    plt.imshow(images[i])

animation.FuncAnimation(fig, update, frames=range(len(images)), interval=200).save('input.gif', writer='pillow', fps=10)

```
